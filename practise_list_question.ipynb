```python
#swap first and last element of a list


def new_list(l):
  size=len(l)
  tem=l[0]
  l[0]=l[size-1]
  l[size-1]=tem

  return l
l=[1,2,3,4,5]
print(new_list(l))
```

```python
#swap entire element of a list
def swap(l):
  size=len(l)
  temp=l[0]
  l[0]=l[1]
  l[1]=temp
  return l
l=[1,2,3,4,5,6]
print(swap(l))
```

```python
#remove common element from two list
l1=[1,2,3]
l2=[3,4,5]
def cm(l1,l2):
  for i in l1[:]:
    for i in l2:
      l1.remove(i)
      l2.remove(i)
```

```python
#convert list to list of dictionary


test_list=["ayush",23,"is",27,"best",10,"for",18,"greeks",33]
print("original list is",str(test_list))

key_list=["name","no"]
res=[]
n=len(test_list)
for i in range(0,n,2):
  res.append({key_list[0]:test_list[i],key_list[1] : test_list[i+1]})

  print("Structured list",str(res))
```
